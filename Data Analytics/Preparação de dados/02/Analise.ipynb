```python
import json

with open("dataset-telecon.json") as f:
    jsonb = json.load(f)
 
datan = pd.json_normalize(jsonb)
datan.head()
```

```python
datan.head()
```

```python
datan.info()
```

```python
#Atribuindo idx para se referir aos valores nulos
idx = datan[datan['conta.cobranca.Total'] == ' '].index

#Pela visualização pode visualizar que todas cobrancas mensais eram de 2 anos, portanto vamos atribuir a todas elas 24
datan.loc[idx, "cliente.tempo_servico"] = 24

#Multiplicação do tempo de servico pelo valor mensal
datan.loc[idx, "conta.cobranca.Total"] = datan.loc[idx, "conta.cobranca.mensal"] *24

# Visualização dos dados alterados
datan.loc[idx][
    ['cliente.tempo_servico', 'conta.contrato', 'conta.cobranca.mensal', 'conta.cobranca.Total']
]
```

```python
datan['conta.cobranca.Total'] = datan['conta.cobranca.Total'].astype(float)
datan.info()
```

```python
for col in datan.columns:
  print(f"Coluna:  {col}")
  print(datan[col].unique())
  print("-"*30)
```
